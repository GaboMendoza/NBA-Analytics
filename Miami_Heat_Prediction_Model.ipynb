## 5.- Multiple Linear Regression

Primero creamos la base de variables que necesitaremos para nuestro modelo:

```python
cdf=MH_games_2020[['PTS','FG_PCT','FG3_PCT','DREB','REB','AST','PLUS_MINUS']]
cdf
```

#### We create our train/test data

```python
y_data=cdf['PLUS_MINUS']
x_data=cdf.drop('PLUS_MINUS',axis=1)
```

Now we randomly split our data into training and testing data  using the function <b>train_test_split</b>. 

```python
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.10, random_state=1)


print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

Now, WE PERFORM OUR MULTIPLE LINEAR REGRESSION MODEL:

```python
from sklearn import linear_model
```

```python
lre=linear_model.LinearRegression()
```

```python
lre.fit(x_train, y_train)
```

```python
## R^2 Para la predicción
lre.score(x_test, y_test)
```

```python
## R^2 Para el entrenamiento
lre.score(x_train, y_train)
```

```python
## Coeficientes
lre.coef_
```

```python
## Cross Validation
from sklearn.model_selection import cross_val_score
```

```python
Rcross=cross_val_score(lre,x_data,y_data,cv=6)
```

```python
Rcross
```

```python
yhat_train=lre.predict(x_train)
yhat_train[0:5]
```

```python
yhat_test=lre.predict(x_test)
yhat_test[0:5]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))

    ax1 = sns.distplot(RedFunction, hist=False, color="r", label=RedName)
    ax2 = sns.distplot(BlueFunction, hist=False, color="b", label=BlueName, ax=ax1)

    plt.title(Title)
    plt.xlabel('PLUS_MINUS')
    plt.ylabel('Model')

    plt.show()
    plt.close()
```

```python
Title = 'Distribution  Plot of  Predicted Value Using Training Data vs Training Data Distribution'
DistributionPlot(y_train, yhat_train, "Actual Values (Train)", "Predicted Values (Train)", Title)
```

```python
Title = 'Distribution  Plot of  Predicted Value Using Training Data vs Training Data Distribution'
DistributionPlot(y_test, yhat_test, "Actual Values (Train)", "Predicted Values (Train)", Title)
```

### Concluimos que no es posible estimar un modelo de regresión lineal para estimar el PLUS_MINUS y por lo tanto la probabilidad de ganar del Miami Heat.

### Igualmente concluimos que los Rebotes, Asistencias, y % FG3_PCT tienen un 48% de relación con el resultado.

## 6.- En esta sección analizaremos qué tipo modelo de regresión puede ayudarnos a entender mejor la relación del PLUS MINUS con cualquier variable que tenemos

```python
import numpy as np
```

```python
msk=np.random.rand(len(cdf))<0.8
train=cdf[msk]
test=cdf[~msk]
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
train_x = np.asanyarray(train[['DREB']])
train_y = np.asanyarray(train[['PLUS_MINUS']])

test_x = np.asanyarray(test[['DREB']])
test_y = np.asanyarray(test[['PLUS_MINUS']])


poly = PolynomialFeatures(degree=2)
train_x_poly = poly.fit_transform(train_x)
train_x_poly

test_x_poly=poly.fit_transform(test_x)
test_x_poly
```

```python
clf = linear_model.LinearRegression()
train_y_ = clf.fit(train_x_poly, train_y)
# The coefficients
print ('Coefficients: ', clf.coef_)
print ('Intercept: ',clf.intercept_)
print('R Score: ',clf.score(test_x_poly,test_y))
```

```python
train_yhat=clf.predict(train_x_poly)
train_yhat[0:5]
```

```python
test_yhat=clf.predict(test_x_poly)
test_yhat[0:5]
```

```python
Title = 'Distribution  Plot of  Predicted Value Using Training Data vs Training Data Distribution'
DistributionPlot(train_y, train_yhat, "Actual Values (Train)", "Predicted Values (Train)", Title)
```

```python
Title = 'Distribution  Plot of  Predicted Value Using Training Data vs Training Data Distribution'
DistributionPlot(test_y, test_yhat, "Actual Values (Train)", "Predicted Values (Train)", Title)
```

```python
Y=np.asarray(MH_games_2020[['Win']])
Y[0:5]
```

NORMALIZE THE DATA SET:

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

### Train/Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.4, random_state=4)
print ('Train set:', X_train.shape,  Y_train.shape)
print ('Test set:', X_test.shape,  Y_test.shape)
```

### Modeling Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='saga').fit(X_train,Y_train)
LR
```

```python
Yhat = LR.predict(X_test)
Yhat
```

```python
Yhat_prob = LR.predict_proba(X_test)
Yhat_prob
```

```python
from sklearn.metrics import log_loss
log_loss(Y_test, Yhat_prob)
```
